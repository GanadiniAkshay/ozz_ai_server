```python
import zipfile
import os
import json
```

```python
ozz_data = {
    "ozz_data":{
        "intents":[],
        "entities":[]
    }
}

intentObjs = []
```

```python
zip_ref = zipfile.ZipFile('./data/dialogflow/EmailIntentTag.zip', 'r')
zip_ref.extractall("./data/dialogflow/contents/")
zip_ref.close()
```

```python
intents = os.listdir('./data/dialogflow/contents/intents')
```

```python
for intent in intents:
    if "_usersays_en.json" in intent:
        intentName = intent.replace("_usersays_en.json","")
        intentObj = {"name":intentName,"utterances":[],"responses":[]}
        utterances = []
        with open('./data/dialogflow/contents/intents/'+intent,"r") as f:
            data = json.load(f)
            for item in data:
                for example in item['data']:
                    utterances.append(example['text'])
            intentObj['utterances'] = utterances
        intentObjs.append(intentObj)
```

```python
ozz_data["ozz_data"]["intents"] = intentObjs
```

```python
with open('./data/dialogflow/rasaScribe.json','w') as outfile:
    json.dump(ozz_data,outfile)
```
